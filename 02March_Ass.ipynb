```python
'''Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
Answer-A scatter plot is a diagram where each value in the data set is represented by a dot. The Matplotlib module has a 
method for drawing scatter plots, it needs two arrays of the same length, one for the values of the x-axis, and one for
the values of the y-axis:'''
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.title("Normal random values graph")
plt.xlabel("These are the values of x")
plt.ylabel("These are the values of y")
```

```python
'''Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Answer-Description. subplot( m , n , p ) divides the current figure into an m -by- n grid and creates axes in the position 
specified by p . MATLAB numbers subplot positions by row. The first subplot is the first column of the first row, the second 
subplot is the second column of the first row, and so on.'''
import numpy as np
#Data for the four lines
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
fig,ax=plt.subplots(2,2)
ax[0,0].plot(x1,y1)
ax[0,0].set_title("Line 1")
ax[0,1].plot(x2,y2)
ax[0,1].set_title("Line 2")
ax[1,0].plot(x3,y3)
ax[1,0].set_title("Line 3")
ax[1,1].plot(x4,y4)
ax[1,1].set_title("Line 4")
```

```python
'''Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
Answer-Bar charts are one of the many techniques used to present data in a visual form so that the reader may readily 
recognize patterns or trends. Bar charts usually present categorical variables, discrete variables or continuous 
variables grouped in class intervals.'''
import numpy as np
import pandas as pd
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.barh(company,profit)
```

```python
'''Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
Answer-When we display the data distribution in a standardized way using 5 summary  minimum, Q1 (First Quartile),
median, Q3(third Quartile), and maximum, it is called a Box plot. It is also termed as box and whisker plot. '''
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
fig,ax=plt.subplots(1,2)
plt.figure(figsize=(5,10))
ax[0].boxplot(box1)
ax[0].set_title("This is boxplot of box1")
ax[1].boxplot(box2)
ax[1].set_title("This is the boxplot of box2")
plt.show()
```
