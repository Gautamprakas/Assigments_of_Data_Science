```python
'''Q1. How can you create a Bokeh plot using Python code?
Answer-first we havw to import all the modules that are required to plot graph in bokeh like import bokeh.io,import bokeh.plotting,bokeh.io.output_notebook()
then we are able to plot the graph in bokeh'''
#In example
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
```

```python
from bokeh.plotting import figure,output_file,show
from bokeh.sampledata.iris import flowers # it import iris dataset as flowers dataframe
output_file("flower.html")
p=figure(title="Flower Dataset")
p.xaxis.axis_label="petal_length"
p.yaxis.axis_label="petal_width"
p.circle(flowers['petal_length'],flowers['petal_width'])
show(p)
```

```python
'''Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.
Answer-Bokeh and Advanced Layouts
Glyphs are the 'basic visual building blocks of Bokeh plots' [1] and refer to vectorised graphical shapes or markers that are used to represent your data.'''
from bokeh.plotting import figure, show
from bokeh.io import output_file

# prepare some data
x = [1, 2, 4, 6, 7]
y = [7, 6, 3, 9, 10]

# create a new plot with figure function
p = figure(title="Circle Glyph", plot_width=450, plot_height=400)

# create circle glyph
p.circle(x=x, y=y, size=25, fill_color="red")
p.circle_cross(x=[2, 4, 6, 8], y=[5, 8, 9, 11], size=25,
			fill_color="blue", fill_alpha=0.3, line_width=2)
p.circle_x(x=[4, 7, 2, 6], y=[7, 2, 4, 9], size=25,
		fill_color="green", fill_alpha=0.6, line_width=2)

# show the results
output_file('circle.html')
show(p)
```

```python
'''Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?'''
output_file("flower.html")
p=figure(title="Flower Dataset")
p.xaxis.axis_label="petal_length"
p.yaxis.axis_label="petal_width"
p.circle(flowers['petal_length'],flowers['petal_width'],color='red',legend_label='red Points',size=15)
show(p)
```
